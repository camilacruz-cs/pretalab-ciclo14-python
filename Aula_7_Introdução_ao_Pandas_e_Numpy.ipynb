### Estruturas do **NumPy**

O NumPy fornece uma estrutura central chamada **ndarray (N-dimensional array)**.  
É um **array homogêneo** (todos os elementos têm o mesmo tipo, geralmente numérico).

#### Características principais:
- Armazena dados de forma **contígua na memória** → operações vetorizadas são muito rápidas.
- É eficiente para **operações matemáticas e estatísticas**.
- Suporta **múltiplas dimensões** (1D, 2D, 3D...).
- Os tipos de dados (`dtype`) são fixos, como `int32`, `float64`, etc.

| Tipo | Estrutura | Exemplo |
|------|------------|----------|
| 1D | Vetor | `[1, 2, 3, 4]` |
| 2D | Matriz | `[[1, 2, 3], [4, 5, 6]]` |
| 3D | Tensor | `[[[1, 2], [3, 4]], [[5, 6], [7, 8]]]` |

```python
#Criando um array a partir de uma lista Python
lista = [1, 2, 3, 4, 5]
arr = np.array(lista)

print("Array criado:", arr)
print("Tipo:", type(arr))
```

```python
#Indexação e Fatiamento
arr = np.array([10, 20, 30, 40, 50])

print("Primeiro elemento:", arr[0])
print("Último elemento:", arr[-1])
print("Fatia do índice 1 ao 3:", arr[1:4])
```

```python
#Operações matemáticas
arr = np.array([10, 20, 30, 40])

print("Array original:", arr)
print("Somando 10 a cada elemento: ", arr + 10)
print("Multiplicando por 2:", arr*2)
print("Dividindo por 5:", arr/5)
```

```python
print("Média:", np.mean(arr) )
print("Máximo:", np.max(arr))
print("Mínimo:", np.min(arr) )
```

## 3. Trabalhando com Pandas

O Pandas é ideal para trabalhar com dados tabulares, semelhantes a planilhas do Excel.
Ele se baseia no NumPy, mas oferece rótulos, colunas nomeadas e funcionalidades de manipulação mais ricas.

https://pandas.pydata.org/docs/

### 3.1 Estruturas do **Pandas**

O Pandas introduz duas estruturas principais:

1. **Series:** equivalente a uma coluna de uma planilha (1D).  
2. **DataFrame:** uma tabela bidimensional, com linhas e colunas nomeadas.

Essas estruturas são construídas **sobre arrays do NumPy**, adicionando:
- **Rótulos** (índices) para linhas e colunas.
- **Tipos mistos** de dados (numéricos, texto, datas, etc.).
- **Métodos de leitura e escrita** (CSV, Excel, SQL, etc.).
- **Ferramentas de análise e agregação.**

### Resumo: Comparação entre NumPy e Pandas

| Característica | NumPy | Pandas |
|----------------|-------|--------|
| **Tipo de dados** | Homogêneo (todos os elementos do mesmo tipo) | Heterogêneo (cada coluna pode ter um tipo diferente) |
| **Estrutura principal** | `ndarray` (array N-dimensional) | `Series` e `DataFrame` |
| **Foco principal** | Cálculos numéricos e operações vetorizadas | Manipulação e análise de dados tabulares |
| **Dimensões** | N-dimensionais (1D, 2D, 3D...) | Principalmente 1D e 2D |
| **Rótulos (índices)** | Acesso posicional (índices numéricos) | Acesso por rótulo (nome de colunas e linhas) |
| **Operações** | Matemáticas, lógicas e estatísticas | Limpeza, filtragem, agrupamento e agregação |
| **Entrada/Saída de dados** | Não possui métodos nativos (usa arrays diretos) | Suporte nativo a CSV, Excel, JSON, SQL, etc. |
| **Performance** | Mais rápido para cálculos numéricos puros | Levemente mais lento, mas muito mais flexível |
| **Uso típico** | Computação científica e vetorização | Análise e manipulação de dados |
